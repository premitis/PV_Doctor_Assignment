```python
import pandas as pd
import glob
import os

def merge_csv_files_from_all_subfolders(parent_folder_path):
    combined_dfs = {}
    parent_folders = [f.path for f in os.scandir(parent_folder_path) if f.is_dir()]

    for parent_folder in parent_folders:
        subfolders = [f.path for f in os.scandir(parent_folder) if f.is_dir()]

        for subfolder in subfolders:
            csv_files = glob.glob(os.path.join(subfolder, '*.csv'))
            
            for csv_file in csv_files:
                df = pd.read_csv(csv_file)
                subfolder_name = os.path.basename(subfolder)
                
                if subfolder_name not in combined_dfs:
                    combined_dfs[subfolder_name] = []
                
                combined_dfs[subfolder_name].append(df)
    
    merged_dfs = {}
    for subfolder_name, dfs in combined_dfs.items():
        if dfs:
            merged_dfs[subfolder_name] = pd.concat(dfs, ignore_index=True)
    
    return merged_dfs

def merge_all_dataframes(merged_dfs):
    final_df = pd.DataFrame()

    for subfolder_name, df in merged_dfs.items():
        if final_df.empty:
            final_df = df
        else:
            final_df = pd.merge(final_df, df, on='Date', how='outer', suffixes=('', f'_{subfolder_name}'))
    
    filtered_df = final_df[['Date', 'GHI', 'PR']]

    for col in final_df.columns:
        if 'GHI_' in col:
            filtered_df['GHI'] = filtered_df['GHI'].fillna(0) + final_df[col].fillna(0)
        elif 'PR_' in col:
            filtered_df['PR'] = filtered_df['PR'].fillna(0) + final_df[col].fillna(0)

    filtered_df = filtered_df.groupby('Date').sum().reset_index()
    
    return filtered_df

parent_folder_path = 'C:\\Users\\Prem\\Desktop\\SEM7\\PLACEMENT\\PV_Doctor_Internship\\data'

merged_dfs = merge_csv_files_from_all_subfolders(parent_folder_path)

final_df = merge_all_dataframes(merged_dfs)

output_path = 'C:\\Users\\Prem\\Desktop\\SEM7\\PLACEMENT\\PV_Doctor_Internship\\data\\cleaned_combined_output.csv'
final_df.to_csv(output_path, index=False)

print(f'Cleaned merged CSV file saved to {output_path}')
```
